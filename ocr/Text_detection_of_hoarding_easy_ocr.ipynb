### Language Hindi 'hi', English 'en', Marathi 'mr'.
#### Download the model

```python
reader = easyocr.Reader(["en"], gpu = False) #,'hi'
```

```python
import PIL
from PIL import ImageDraw
im = PIL.Image.open("C:/Users/Abir Khan/Desktop/crop/1.jpg")
im
```

### Doing OCR & Get bounding boxes.

```python
bounds = reader.readtext('C:/Users/Abir Khan/Desktop/crop/1.jpg')
bounds
```

### Draw bounding boxes

```python
def draw_boxes(image, bounds, color='yellow', width=8):
    draw = ImageDraw.Draw(image)
    for bound in bounds:
        p0, p1, p2, p3 = bound[0]
        draw.line([*p0, *p1, *p2, *p3], fill=color, width=width)
    return image
draw_boxes(im, bounds)
        
```
